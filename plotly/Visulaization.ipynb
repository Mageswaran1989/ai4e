### Taylor Series
- https://tutorial.math.lamar.edu/classes/calcii/taylorseries.aspx

```python
def f(x):
    return math.cos(x)

X = np.linspace(-10,10,1000)
Y = [f(x) for x in X]

fig = px.line({"x": X, "y":Y}, x="x", y="y", title='cosx)')
fig.show()
```

```python
def f_derivative_1(x):
    return -math.sin(x)

X = np.linspace(-10,10,1000)
Y = [f_derivative_1(x) for x in X]

fig = px.line({"x": X, "y":Y}, x="x", y="y", title='cosx)')
fig.show()
```

```python
def f_degree2(x):
    # Plot degree 2 Taylor polynomial 
    return 1 - x**2 / 2

X = np.linspace(-10,10,1000)
Y = [f_degree2(x) for x in X]

fig = px.line({"x": X, "y":Y}, x="x", y="y", title='cosx)')
fig.show()
```

```python
def f_degree4(x):
    # Plot degree 4 Taylor polynomial 
    return 1 - x**2 / 2 + x**4 / 24

X = np.linspace(-10,10,1000)
Y = [f_degree4(x) for x in X]

fig = px.line({"x": X, "y":Y}, x="x", y="y", title='cosx)')
fig.show()
```

```python
fig = go.Figure()
X = np.linspace(-6,6,50)

Y = np.cos(X)
fig.add_trace(go.Scatter(x=X, y=Y,
                    mode='lines',
                    name='cos(x)'))

Y2 = 1 - X**2 / 2
fig.add_trace(go.Scatter(x=X, y=Y2,
                    mode='lines',
                    name='Degree2'))
Y4 = 1 - X**2 / 2 + X**4 / 24
fig.add_trace(go.Scatter(x=X, y=Y4,
                    mode='lines',
                    name='Degree4'))

fig.update_layout(
    xaxis=dict(range=[-6,6]),
    yaxis=dict(range=[-4,4]))
        
fig.show()
```

```python

# Import libraries 
import matplotlib.pyplot as plt 
import numpy as np 
  
x = np.linspace(-6, 6, 50) 
  
fig = plt.figure(figsize = (14, 8)) 
  
# Plot y = cos(x) 
y = np.cos(x) 
plt.plot(x, y, 'b', label ='cos(x)') 
  
# Plot degree 2 Taylor polynomial 
y2 = 1 - x**2 / 2
plt.plot(x, y2, 'r-.', label ='Degree 2') 
  
# Plot degree 4 Taylor polynomial 
y4 = 1 - x**2 / 2 + x**4 / 24
plt.plot(x, y4, 'g:', label ='Degree 4') 
  
# Add features to our figure 
plt.legend() 
plt.grid(True, linestyle =':') 
plt.xlim([-6, 6]) 
plt.ylim([-4, 4]) 
  
plt.title('Taylor Polynomials of cos(x) at x = 0') 
plt.xlabel('x-axis') 
plt.ylabel('y-axis') 
  
# Show plot 
plt.show() 
```
